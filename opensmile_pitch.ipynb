```python
%matplotlib inline
import ml.audio_feature_extraction
import ml.audio_instance
import ml.signal_processing

import matplotlib.pyplot as plt
from IPython.display import Audio, Image

from pylab import rcParams
rcParams['figure.figsize'] = 16, 5
```

```python
extractor = ml.audio_feature_extraction.AcousticsExtractor(config="audio_features.config")
fn = "charla.wav"

instance = ml.audio_instance.AudioInstance(filename=fn, speaker_gender="m", audio=ml.signal_processing.load_wav(fn))
times, pitch = extractor.extract_pitch(instance)
plt.plot(times, pitch, "*")
plt.ylabel("Freq (Hz)")
plt.ylim([75, 300])
```
